```python
import cv2
import time
# Same command function as streaming, its just now we pass in the file path, nice!
cap = cv2.VideoCapture('MVI_1790.mp4')

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fps = 30

bus_cascade = cv2.CascadeClassifier('required_files/bus.xml')
car_cascade = cv2.CascadeClassifier('required_files/cars3.xml')
bike_cascade = cv2.CascadeClassifier('required_files/bike.xml')
pedestrian_cascade = cv2.CascadeClassifier('required_files/pedestrian.xml')


def detect_road(img):
    face_img = img.copy()
    
    bus_rects = bus_cascade.detectMultiScale(face_img) 
    for (x,y,w,h) in bus_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (0,0,255), 2) 
        cv2.putText(face_img,text='otobus',
                    org=(x,y-6), fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                    fontScale= 0.8, color=(0,0,255),
                    thickness=1, lineType=cv2.LINE_AA)

    car_rects = car_cascade.detectMultiScale(face_img) 
    for (x,y,w,h) in car_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,0,0), 2) 
        cv2.putText(face_img,text='araba',
                    org=(x,y-6), fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                    fontScale= 0.8, color=(255,0,0),
                    thickness=1, lineType=cv2.LINE_AA)

    bike_rects = bike_cascade.detectMultiScale(face_img) 
    for (x,y,w,h) in bike_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (0,255,0), 2) 
        cv2.putText(face_img,text='bisiklet',
                    org=(x,y-6), fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                    fontScale= 0.8, color=(0,255,0),
                    thickness=1, lineType=cv2.LINE_AA)
        
    pedestrian_rects = pedestrian_cascade.detectMultiScale(face_img) 
    for (x,y,w,h) in pedestrian_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 2) 
        cv2.putText(face_img,text='yaya',
                    org=(x,y-6), fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                    fontScale= 0.4, color=(255,255,255),
                    thickness=1, lineType=cv2.LINE_AA)
        
    return face_img

# MACOS AND LINUX: *'XVID' (MacOS users may want to try VIDX as well just in case)
# WINDOWS *'VIDX'
writer = cv2.VideoWriter('result_video.mp4', cv2.VideoWriter_fourcc(*'VIDX'),25, (width, height))


if cap.isOpened()== False: 
    print("Error opening the video file.")
    

while cap.isOpened():
    
    # Read the video file.
    ret, frame = cap.read()
    
    # If we got frames, show them.
    if ret == True:

        # Display the frame at same frame rate of recording
        # Watch lecture video for full explanation
        #time.sleep(1/fps)
        frame = detect_road(frame)
        cv2.imshow('frame',frame)
        writer.write(frame)
 
        # Press q to quit
        if cv2.waitKey(25) & 0xFF == ord('q'):
            
            break
 
    # Or automatically break this whole loop if the video is over.
    else:
        break
        
cap.release()
writer.release()
cv2.destroyAllWindows()
```
